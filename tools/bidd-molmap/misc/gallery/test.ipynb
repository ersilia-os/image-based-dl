```python
import molmap
```

```python
ftypes = ['descriptor', 'fingerprint']
methods = ['umap', 'tsne', 'mds']
metrics = ['cosine', 'correlation']
```

```python
for ftype in ftypes:
    for method in methods:
        for metric in metrics:
            mp = molmap.MolMap(ftype = ftype, metric= metric) 
            mp.fit(method = method)
            mp.plot_scatter(radius=3)
            mp.plot_grid()            
```
