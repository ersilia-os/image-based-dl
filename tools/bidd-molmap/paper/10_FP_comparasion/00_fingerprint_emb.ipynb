```python
import molmap
import os
```

```python
data_save_folder = '/raid/shenwanxiang/FP_maps'
```

```python
ls -lh /raid/shenwanxiang/FP_maps
```

```python
if not os.path.exists(data_save_folder):
    os.makedirs(data_save_folder)
```

```python
metric = 'cosine'
method = 'umap'
n_neighbors = 30
min_dist = 0.1
```

```python
bitsinfo = molmap.feature.fingerprint.Extraction().bitsinfo
bitsinfo.Subtypes.unique()
```

```python
for fp in bitsinfo.Subtypes.unique():
    
    flist = bitsinfo[bitsinfo.Subtypes.isin([fp])].IDs.tolist()
    mp2 = molmap.MolMap(ftype = 'fingerprint', metric = metric, flist =flist,)
    mp2.fit(method = method, n_neighbors = n_neighbors, min_dist = min_dist)    
    mp2.save(os.path.join(data_save_folder, '%s.mp' % fp))
    
    mp2.plot_scatter(htmlname = fp, htmlpath = './html')
    mp2.plot_grid(htmlname = fp, htmlpath = './html')
```
