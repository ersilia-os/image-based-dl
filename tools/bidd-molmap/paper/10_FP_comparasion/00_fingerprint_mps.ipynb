```python
from molmap import show, loadmap, feature
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
sns.set(style='white', font_scale = 2)

size = 20
```

```python
cms = feature.fingerprint.colormaps
```

```python
data_save_folder = '/raid/shenwanxiang/FP_maps'
```

```python
aspirin = 'CC(=O)OC1=CC=CC=C1C(O)=O' #aspirin
NAC = 'CC(=O)NC1=CC=CC=C1C(O)=O' #N_acetylanthranilic_acid
smiles_list = [aspirin, NAC]

fps1 = ['AtomPairFP', 'TorsionFP', 'MorganFP',  'MAP4', 'RDkitFP', 'MHFP6']
fps2 = ['EstateFP', 'MACCSFP', 'PharmacoErGFP', 'PharmacoPFP', 'PubChemFP', 'AvalonFP', ]

fps = fps2


X = []
for fp in fps:  
    mp = loadmap(os.path.join(data_save_folder, '%s.mp' % fp))
    x = mp.batch_transform(smiles_list)
    X.append(x)
    

fig, axes = plt.subplots(nrows=2, ncols=len(fps),  figsize=(20,7))

for i, axs in enumerate(axes):
    for fp, ax, x in zip(fps, axs, X):
        show.imshow(x[i], ax = ax, mode='light', x_max = 1, color_list = [cms.get(fp)])
        ax.set_xlabel(fp, fontsize = size) 
        
for ax, row in zip(axes[:,0], ['Aspirin', 'Analog']):
    ax.set_ylabel(row.capitalize(), fontsize = size) 
fig.tight_layout()
fig.savefig('./html/fp2.svg', bbox_inches='tight', dpi=300)
```
