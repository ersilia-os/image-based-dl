## In case you need to tune the hyperparameter about the feature map object, you can rearrange your feature values X by the rearrangement method, you don't need to extract the feature values again

```python
import molmap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mp_cosine = molmap.loadmap('./descriptor.mp')
mp_tsne = molmap.loadmap('./descriptor.mp')
```

```python
mp_tsne.fit(method='tsne')
```

```python
mp_tsne.plot_scatter('images/')
```

```python
mp_tsne.plot_grid('images/')
```

```python
X0 = mp_cosine.batch_transform(['CC','CCCC','CCCCC'])
X1 = mp_tsne.batch_transform(['CC','CCCC','CCCCC'])
```

## rearrangement

```python
X2 = mp_cosine.rearrangement(X0, mp_tsne)
```

```python
plt.imshow(X0[0].sum(axis=-1))
```

```python
plt.imshow(X1[0].sum(axis=-1))
```

```python
plt.imshow(X2[0].sum(axis=-1))
```

```python
a = X2 != X1
a.sum() #exactly same result
```
