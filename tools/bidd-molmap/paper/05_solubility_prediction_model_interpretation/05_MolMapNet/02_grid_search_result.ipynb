```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
log_file = 'ESOL_Wed_Jan_29_11-02-29_2020.log'
df = pd.read_csv(log_file)
n_neighbors_list = df.n_neighbors.unique()
min_dist_list = df.min_dist.unique()
x = df.valid_best_rmse.values.reshape(len(n_neighbors_list), len(min_dist_list))
dfr = pd.DataFrame(x, columns = min_dist_list, index = n_neighbors_list).round(3)
fig, ax = plt.subplots(figsize = (12,10))
sns.set(font_scale = 1.5)
sns.heatmap(dfr, cmap='rainbow',
                 #vmin = 0.21,  vmax=0.32,
                 square=True, 
                 annot = True, fmt = '.2f',
                 ax = ax,
                 #linewidths = 0.05, linecolor= 'white',
                 alpha = 0.9,
                 annot_kws = {'size': 9},
                 cbar_kws = {'fraction':0.046, 'pad': 0.02,  'label': 'RMSE of The Validation Set'}
                )
ax.set_xlabel('min_dist', size = 17)
ax.set_ylabel('n_neighbors', size = 17)
plt.savefig(log_file + '.png', dpi = 400, bbox_inches ='tight')
```
